```python
# ======================
# COVID-19 DATA TRACKER - COMPLETE WORKING CODE
# ======================

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 1. LOAD AND CLEAN DATA
# ----------------------
try:
    df = pd.read_csv('owid-covid-data.csv')
    print("✅ Data loaded successfully!")
    
    # Data cleaning
    df['date'] = pd.to_datetime(df['date'])
    df = df[df['continent'].notna()]  # Remove continent aggregates
    df = df.sort_values(['location', 'date'])
    
except Exception as e:
    print(f"❌ Error: {e}")

# 2. GLOBAL TRENDS
# ----------------------
plt.figure(figsize=(12, 6))
df.groupby('date')['new_cases_smoothed'].sum().plot(
    title='Global Daily New Cases (Smoothed)',
    color='red',
    grid=True
)
plt.ylabel('Cases')
plt.show()

# 3. COUNTRY COMPARISONS
# ----------------------
plt.figure(figsize=(10, 5))
df.groupby('location')['total_cases'].max().nlargest(10).plot(
    kind='bar',
    color='skyblue',
    title='Top 10 Countries by Total Cases'
)
plt.xticks(rotation=45)
plt.show()

# 4. VACCINATION PROGRESS
# ----------------------
vaccination = df.dropna(subset=['people_vaccinated']).sort_values('date').groupby('location').last()['people_vaccinated'].nlargest(10)
vaccination.plot(
    kind='barh',
    title='Top 10 Vaccinated Countries',
    figsize=(10, 6),
    color='green'
)
plt.xlabel('People Vaccinated')
plt.show()

# 5. FATALITY RATE ANALYSIS (FIXED VERSION)
# ----------------------
df['fatality_rate'] = (df['total_deaths'] / df['total_cases']) * 100
high_case_countries = df[df['total_cases'] > 1000].groupby('location')['fatality_rate'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
high_case_countries.plot(
    kind='bar',
    title='Highest Fatality Rates (Countries with >1000 Cases)',
    color='purple'
)
plt.ylabel('Fatality Rate (%)')
plt.show()

print("✨ All analyses completed successfully!")
```
